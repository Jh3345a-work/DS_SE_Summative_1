```python
import tkinter as tk
import matplotlib.pyplot as plt
import pandas as pd

root = tk.Tk()
root.title("Population Estimates Visualisation")
root.geometry("800x600")

# Containers
container = tk.Frame(root)
container.pack(fill="both", expand=True)
container.grid_rowconfigure(0, weight=1)
container.grid_columnconfigure(0, weight=1)

# Pages
landing_page = tk.Frame(container)
timeseries_page = tk.Frame(container)
bar_chart_page = tk.Frame(container)
pie_chart_page = tk.Frame(container)

for page in (landing_page, timeseries_page, bar_chart_page, pie_chart_page):
    page.grid(row=0, column=0, sticky="nsew")
    page.grid_columnconfigure(0, weight=1)

# Homepage content
title = tk.Label(landing_page, text="NOMIS Population Estimates", font=("Arial", 25))
title.grid(row=0, column=0, pady=30, sticky="n")

tk.Button(landing_page, text="Timeseries Graph",
          command=lambda: timeseries_page.tkraise()).grid(row=1, column=0, pady=10)

tk.Button(landing_page, text="Bar Chart",
          command=lambda: bar_chart_page.tkraise()).grid(row=2, column=0, pady=10)

tk.Button(landing_page, text="Pie Chart",
          command=lambda: pie_chart_page.tkraise()).grid(row=3, column=0, pady=10)

tk.Button(landing_page, text="Exit App", command=root.destroy).grid(row=4, column=0, pady=30)

# Labels and Buttons
tk.Label(bar_chart_page, text="Population Estimate per Region").pack(pady=20)
tk.Button(bar_chart_page, text="Back", command=lambda: landing_page.tkraise()).pack(pady=10)


# Timeseries Page
tk.Label(timeseries_page, text="Timeseries Placeholder").pack(pady=20)
tk.Button(timeseries_page, text="Back", command=lambda: landing_page.tkraise()).pack()


# Pie Chart Page
tk.Label(pie_chart_page, text="Pie Chart Placeholder").pack(pady=20)
tk.Button(pie_chart_page, text="Back", command=lambda: landing_page.tkraise()).pack()

# Start
landing_page.tkraise()
```
